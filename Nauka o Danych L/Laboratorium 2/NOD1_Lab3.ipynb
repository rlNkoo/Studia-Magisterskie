## 3.1 Rodzaje wykresów podstawowych

### 3.1.1 Wykres Liniowy

```python
dane = df['release_year'].value_counts().sort_index()
plt.plot(dane.index, dane.values, marker='o', color='blue')
plt.title("Liczba filmów Netflixa w poszczególnych latach")
plt.xlabel("Rok wydania")
plt.ylabel("Liczba tytułów")
plt.show()
```

### 3.1.2 Wykres słupkowy

```python
kraje = df['country'].value_counts().head(10)

plt.bar(kraje.index, kraje.values, color='orange')
plt.title("Top 10 krajów z największą liczbą tytułów")
plt.xlabel("Kraj")
plt.ylabel("Liczba tytułów")
plt.xticks(rotation=45)
plt.show()
```

### 3.1.3 Histogram

```python
plt.hist(df['release_year'], bins=20, color='green', edgecolor='black')
plt.title("Histogram – liczba produkcji Netflixa w kolejnych latach")
plt.xlabel("Rok")
plt.ylabel("Liczba tytułów")
plt.show()
```

### 3.1.4 Wykres kołowy

```python
typy = df['type'].value_counts()

plt.pie(typy.values, labels=typy.index, autopct='%1.1f%%', colors=['#66b3ff','#ff9999'])
plt.title("Udział filmów i seriali w katalogu Netflixa")
plt.show()
```

## 3.2 Zaawansowane typy wykresów

### 3.2.1 Wykres punktowy (scatter plot)

```python
x = np.random.rand(50)
y = np.random.rand(50)
sizes = 500 * np.random.rand(50)
colors = np.random.rand(50)

plt.scatter(x, y, s=sizes, c=colors, alpha=0.5, cmap="viridis")
plt.colorbar()
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Wykres Punktowy")
plt.show()
```

### 3.2.2 Wykres 3D

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()

ax = fig.add_subplot(111, projection="3d")

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

ax.plot_surface(X, Y, Z, cmap="viridis")
plt.title("Wykres 3D")
plt.show()
```

## 3.3 Bibilioteka Plotnine

```python
from plotnine import *
import pandas as pd
import numpy as np
```

```python
data = pd.DataFrame({
 "x": [3, 3.25, 2.75, 3, 2.5],
 "y": [2, 6, 6, 10, 6]
 })
```

```python
ggplot(data) + aes(x="x", y="y") + geom_line(color="red", linetype="dashed") + ggtitle("Wykres testowy") + xlab("X") + ylab("Y")
```

### 3.3.1 Wykres punktowy

```python
ggplot(data) + aes(x="x", y="y") + geom_point() + ggtitle("Wykres Punktowy") + xlab("X") + ylab("Y")
```

### 3.3.2 Wykres punktowy

```python
data_bar = pd.DataFrame({
 "kategorie": ["A", "B", "C", "D"],
 "wartości": [4, 7, 1, 8]
 })
```

```python
ggplot(data_bar) + aes(x="kategorie", y="wartości") + geom_bar(stat="identity", fill="#123456", color="black") + ggtitle("Wykres słupkowy") + xlab("Kategoria") + ylab("Wartość")
```

### 3.3.3 Histogram

```python
data_hist = pd.DataFrame({
    "wartosci": np.random.normal(0, 100_000_000, 500)
})

ggplot(data_hist) + aes(x="wartosci") + geom_histogram(bins=30, fill="#123321") + xlab("Wartość") + ylab("Częstotliwość występowania") + ggtitle("Histogram")
```

### 3.3.4 Linia trendu

```python
const = 5

data_trend = pd.DataFrame({
  "x" : np.linspace(0, 10, 100),
  "y" : (2 * const + 1 + np.random.normal(0, 2, 100))
})

ggplot(data_trend) + aes(x="x", y="y") + geom_point() + geom_smooth() + ggtitle("Linia trendu") + xlab("X") + ylab("Y")
```

### 3.3.5 Facetowanie danych ( podział na podgrupy )

```python
data_facet = pd.DataFrame({
 "x": np.tile([1, 2, 3, 4], 2),
 "y": [1, 2, 3, 4, 2, 3, 4, 5],
 "grupa": ["A"]*4 + ["B"]*4
 })

ggplot(data_facet) + aes(x="x", y="y") + geom_point() + facet_wrap("grupa") + ggtitle("Facetowanie danych") + xlab("X") + ylab("Y") + theme_bw()
```

## 3.4 Wykresy przestrzenne

### 3.4.1 Mapy punktowe

```python
import plotly.express as px
import pandas as pd
```

```python
data_map = pd.DataFrame({
  "city": ["New York", "London", "Tokyo", "Sydney"],
  "latitude": [40.7128, 51.5074, 35.6895,-33.8688],
  "longitude": [-74.0060,-0.1278, 139.6917, 151.2093]
})
```

```python
fig = px.scatter_mapbox(data_map, lat="latitude", lon="longitude", hover_name="city",zoom=1.75, height=1000)
fig.update_layout(mapbox_style="open-street-map")
fig.update_layout(title="Lokalizacja Wybranych Miast na Mapie")
fig.show()
```

### 3.4.2 Mapy o zmiennym kolorze wartości ( Choropleth )

```python
import plotly.express as px
```

```python
data = px.data.gapminder().query("year == 2007")

fig = px.choropleth(data, locations="iso_alpha", color="pop", hover_name="country", color_continuous_scale="Viridis",  labels={"pop": "Populacja"})

fig.update_layout(title="Populacja Kraj w na wiecie w 2007 roku")
fig.show()
```
